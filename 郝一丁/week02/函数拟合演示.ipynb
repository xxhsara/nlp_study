```python
from  import train_sin_fitting_model
import matplotlib.pyplot as plt
import numpy as np

#  sin(x)
X, y_true, y_pred, losses = train_sin_fitting_model(
    num_samples=150,
    x_range=(-2 * np.pi, 2 * np.pi),
    hidden_layers=[32, 32],
    lr=0.01,
    num_epochs=1000
)

# 
plt.figure(figsize=(8, 5))
plt.scatter(X, y_true, color='blue', alpha=0.6, label='Data (sin(x))')
sorted_idx = np.argsort(X.flatten())
plt.plot(X[sorted_idx], y_pred[sorted_idx], color='red', linewidth=2, label='MLP Prediction')
plt.title("Fitting sin(x) with a 2-Layer Neural Network")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

#  loss
# print(f"Final loss: {losses[-1]:.6f}")
```
