```python
import matplotlib.pyplot as plt
from 深度学习分类 import train_and_evaluate_model
DATASET_PATH = "dataset.csv"

# 定义不同模型结构
configs = [
    {"name": "1-64", "hidden_dims": [64]},
    {"name": "1-128", "hidden_dims": [128]},
    {"name": "2-128_64", "hidden_dims": [128, 64]},
    {"name": "2-256_128", "hidden_dims": [256, 128]},
]

results = {}

# 训练各模型
for config in configs:
    print(f"训练模型: {config['name']}")
    losses = train_and_evaluate_model(
        dataset_path=DATASET_PATH,
        hidden_dims=config["hidden_dims"],
        lr=0.01,
        batch_size=32,
        num_epochs=10,
        max_len=40,
        optimizer_type="SGD"
    )
    results[config["name"]] = losses

# 绘制 loss 曲线
plt.figure(figsize=(12, 6))
for name, losses in results.items():
    plt.plot(losses, label=name)
plt.title("training loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
